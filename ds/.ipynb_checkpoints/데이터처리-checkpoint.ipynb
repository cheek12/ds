```python
pos_dic = {'FW':[0,4], 'AM(L)':[2,3],'AM(CL)':[1,3],'AM(C)':[0,3],'AM(CR)':[1,3], 'AM(R)':[2,3],
            'M(L)':[2,2], 'M(CL)':[1,2],'M(C)':[0,2],'M(CR)':[1,2],'M(R)':[2,2], 'DMC':[0,1],
            'D(L)':[2,0], 'D(CL)':[1,0], 'D(C)':[0,0], 'D(CR)':[1,0],'D(R)':[2,0], 
           'AM(CLR)':[1,3],'M(CLR)':[1,2],'D(CLR)':[1,0],
            'AM(LR)' :[2,3],'M(LR)':[2,2], 'D(LR)':[2,0]}
```

```python
pos_dic['FW']
```

```python
refined = refined[refined['Position'] != 'GK']
```

```python
refined['Position'].replace('Midfielder','M(C)',inplace=True)
```

```python
refined
```

```python
refined['Position'].replace('Forward','FW',inplace=True)
```

```python
refined['Position'].replace('Defender','D(C)',inplace=True)
```

```python
sorted(refined['Position'].unique())
```

```python
refined.reset_index(inplace=True)
refined.drop(['index'],axis=1,inplace=True)
```

```python
refined
```

```python
pos_xlist=[0*i for i in range(len(refined['Position']))]
pos_ylist=[0*i for i in range(len(refined['Position']))]
for i in range(len(refined['Position'])):
    count = 0
    correction_x=False # FW는 측면 정보가 없어서 x좌표에 보정 늘거야, FW와 다른 포지션 같이 있는 데이터만
    for posdic in pos_dic.keys():
        if refined['Position'][i].find(posdic) != -1:
            if posdic == 'FW':
                correction_x = True
            pos_xlist[i] += pos_dic[posdic][0]
            pos_ylist[i] += pos_dic[posdic][1]
            count+=1
    if count == 0:
        print(refined['Position'][i])
    
    if correction_x and count != 1:
        pos_xlist[i] += pos_xlist[i]/(count-1) # FW x값 보정, FW좌표를 뺀 나머지 포지션의 좌표 평균을 x에 더해줌
    pos_xlist[i] = pos_xlist[i]/count
    pos_ylist[i] = pos_ylist[i]/count
```

```python
position_df = pd.DataFrame(zip(pos_xlist,pos_ylist), columns=['posX','posY'])
position_df
```

```python
position_df.plot.scatter(x='posX',y='posY') # 포지션을좌표로 나타낸거
```

```python
pos_test = pd.concat([refined['Position'],position_df],axis=1)
pos_test.plot.scatter(x='posX',y='posY')
```
