```python
import pandas as pd
import numpy as np
```

```python
ligue = pd.read_csv('./dataset/refinedDataSet_withPos.csv',index_col=0)
champs = pd.read_csv('./dataset/Europe-Champions-League-2021-2022.csv' ,index_col=0)
```

```python
champs.reset_index(inplace=True)
champs.drop(['index'],axis=1,inplace=True)
champs['Age']=np.nan
for i in range(len(champs)):
    champs.loc[i,'Age']=champs.loc[i,'Player.1'].split()[-2].replace(',','')
    
champs
```

```python
champs['Age']=champs['Age'].astype(float)
champs
```

ㅇㅇㅇ


```python
#여기서 붙일 데이터프레임 column 미리 nan값으로 넣기 -> 안넣어진 row dropna로 한번에 없에기

ligue['Champs_Rating'] = np.nan
ligue['Champs_player']=np.nan

ligue
```

```python

for ch_index in range(len(champs)):
    for index in range(len(ligue)):
        if champs.loc[ch_index,'Player.1'].find(ligue.loc[index,'Name']) != -1 and champs.loc[ch_index,'Age'] == ligue.loc[index,'Age']:
            # 이름이랑 나이로 같은 선수인지 확인 후 값 넣기
            ligue.loc[index,'Champs_Rating'] = champs.loc[ch_index,'Rating']
            ligue.loc[index,'Champs_player'] = champs.loc[ch_index,'Player.1']
            
```
